```python
from transformers import AutoModel, AutoTokenizer
```

```python
tokenize = AutoTokenizer.from_pretrained("microsoft/deberta-base")
```

```python
model = AutoModel.from_pretrained("microsoft/deberta-v3-xsmall")
```

```python
tokens = tokenize('Hello world', return_tensors='pt')
```

```python
output = model(**tokens)[0]
```

```python
output.shape
```

```python
for token in tokens['input_ids'][0]:
  print(tokenize.decode(token))
```
