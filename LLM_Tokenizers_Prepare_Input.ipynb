```python
from transformers import AutoModelForCausalLM, AutoTokenizer
```

```python
model = AutoModelForCausalLM.from_pretrained("microsoft/Phi-3-mini-4k-instruct", device_map="cuda", torch_dtype="auto", trust_remote_code=True)
```

```python
from transformers import AutoConfig

config = AutoConfig.from_pretrained("microsoft/Phi-3-mini-4k-instruct")

# Rename the key if it exists under a different name
if "rope_scaling" in config.__dict__ and "type" not in config.rope_scaling:
    # Common alternative key names
    config.rope_scaling["type"] = config.rope_scaling.get("rope_type", "longrope")

model = AutoModelForCausalLM.from_pretrained("microsoft/Phi-3-mini-4k-instruct", config=config)
```

```python
#Load Tokenizer
tokenizer = AutoTokenizer.from_pretrained("microsoft/Phi-3-mini-4k-instruct")
```

```python
#Define the prompts
prompt = "Write an email apologizing to Sarah for the tragic gardening mishap. Explain how it happend.<|assistant|>"
```

```python
#Tokenize the input prompt
import torch

if torch.cuda.is_available():
    device = "cuda"
    print("CUDA is available! Using GPU for tokenization.")
else:
    device = "cpu"
    print("CUDA is not available. Falling back to CPU for tokenization. Please ensure your Colab runtime type is set to GPU for better performance.")

input_ids = tokenizer(prompt, return_tensors="pt").input_ids.to(device)
```

```python
#Generate the text
input_ids = input_ids.to('cpu')
generation_output = model.generate(input_ids=input_ids, max_new_tokens=20)
```

```python
print(tokenizer.decode(generation_output[0]))
```
